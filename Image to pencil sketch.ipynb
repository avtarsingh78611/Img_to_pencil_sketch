# Convert normal image to pencil sketch image

```python
#Import the library to get all the modules
import cv2
from PIL import Image 
```

```python
#Read the image from your computer
image = cv2.imread("image1.jpg")

#The next step is to convert the image to  grayscale image
grey_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

display(Image.fromarray(grey_img))
```

```python
#Now we have to invert the image 
```

```python
invert =cv2.bitwise_not(grey_img)
display(Image.fromarray(invert))
```

```python
#in this step we will add the inverted image and blur image
blur = cv2.GaussianBlur(invert, ( 21, 21), 0)
display(Image.fromarray(blur))
```

```python
#We will now mix the grascale image with the inverted blurry image
invertedblur = cv2.bitwise_not(blur)
sketch = cv2.divide(grey_img, invertedblur, scale = 256.0)
display(Image.fromarray(sketch))
```
